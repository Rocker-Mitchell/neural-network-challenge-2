## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder, StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Determine the number of unique values in each column.
attrition_df.nunique()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[["Attrition", "Department"]].copy()
```

```python
# Create a list of at least 10 column names to use as X data
X_columns = [
    "Age",
    "DistanceFromHome",
    "Education",
    "EducationField",
    "EnvironmentSatisfaction",
    "HourlyRate",
    "JobSatisfaction",
    "MaritalStatus",
    "NumCompaniesWorked",
    "OverTime",
    "PerformanceRating",
    "TotalWorkingYears",
    "WorkLifeBalance",
    "YearsAtCompany",
    "YearsSinceLastPromotion",
    "YearsWithCurrManager"
]

# Create X_df using your selected columns
X_df = attrition_df[X_columns].copy()

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_df, y_df, random_state=1)
```

```python
# Convert your X data to numeric data types however you see fit
# Add new code cells as necessary

# Set up DataFrames to hold encodings
X_train_enc = X_train.copy()
X_test_enc = X_test.copy()
```

```python
# Encode EducationField
display(X_train["EducationField"].value_counts())

education_field_encoder = OneHotEncoder(sparse_output=False, dtype="int64")
education_field_encoder.fit(X_train[["EducationField"]])
education_field_columns = education_field_encoder.get_feature_names_out(["EducationField"])

def encode_education_field(X: pd.DataFrame):
    if "EducationField" in X.columns:
        X[education_field_columns] = education_field_encoder.transform(X[["EducationField"]])
        X.drop(columns="EducationField", inplace=True)

encode_education_field(X_train_enc)
encode_education_field(X_test_enc)

X_train_enc[education_field_columns].head()
```

```python
# Encode MaritalStatus
display(X_train["MaritalStatus"].value_counts())

marital_status_encoder = OneHotEncoder(sparse_output=False, dtype="int64")
marital_status_encoder.fit(X_train[["MaritalStatus"]])
marital_status_columns = marital_status_encoder.get_feature_names_out(["MaritalStatus"])

def encode_marital_status(X: pd.DataFrame):
    if "MaritalStatus" in X.columns:
        X[marital_status_columns] = marital_status_encoder.transform(X[["MaritalStatus"]])
        X.drop(columns="MaritalStatus", inplace=True)

encode_marital_status(X_train_enc)
encode_marital_status(X_test_enc)

X_train_enc[marital_status_columns].head()
```

```python
# Encode OverTime
display(X_train["OverTime"].value_counts())

overtime_encoder = OneHotEncoder(drop="first", sparse_output=False, dtype="int64")
overtime_encoder.fit(X_train[["OverTime"]])

def encode_overtime(X: pd.DataFrame):
    if "OverTime" in X.columns and pd.api.types.is_string_dtype(X["OverTime"]):
        X["OverTime"] = overtime_encoder.transform(X[["OverTime"]])

encode_overtime(X_train_enc)
encode_overtime(X_test_enc)

X_train_enc[["OverTime"]].head(10)
```

```python
# Check final encoded X
display(X_train_enc.dtypes)
display(X_train_enc.head())
```

```python
# Create a StandardScaler
scaler = StandardScaler()

# Fit the StandardScaler to the training data
scaler.fit(X_train_enc)

# Scale the training and testing data
X_train_scaled = scaler.transform(X_train_enc)
X_test_scaled = scaler.transform(X_test_enc)
```

```python
y_train["Department"].value_counts()
```

```python
# Create a OneHotEncoder for the Department column
department_encoder = OneHotEncoder(sparse_output=False, dtype="int64")

# Fit the encoder to the training data
department_encoder.fit(y_train[["Department"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_department = department_encoder.transform(y_train[["Department"]])
y_test_department = department_encoder.transform(y_test[["Department"]])

y_train_department
```

```python
y_train["Attrition"].value_counts()
```

```python
# Create a OneHotEncoder for the Attrition column
attrition_encoder = OneHotEncoder(sparse_output=False, dtype="int64")

# Fit the encoder to the training data
department_encoder.fit(y_train[["Attrition"]])

# Create two new variables by applying the encoder
# to the training and testing data
y_train_attrition = department_encoder.transform(y_train[["Attrition"]])
y_test_attrition = department_encoder.transform(y_test[["Attrition"]])

y_train_attrition
```

## Part 2: Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
X_col_len = X_train_scaled.shape[1]

# Create the input layer
input_layer = layers.Input((X_col_len,))

# Create at least two shared layers
dense1_layer = layers.Dense(110, "relu")(input_layer)
dense2_layer = layers.Dense(60, "relu")(dense1_layer)
dense3_layer = layers.Dense(45, "relu")(dense2_layer)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer

# Create the hidden layer
department_dense_layer = layers.Dense(30, "relu")(dense3_layer)

# Create the output layer
department_output_layer = layers.Dense(y_train_department.shape[1], "sigmoid", name="department_output")(department_dense_layer)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attrition_dense_layer = layers.Dense(25, "relu")(dense3_layer)

# Create the output layer
attrition_output_layer = layers.Dense(y_train_attrition.shape[1], "softmax", name="attrition_output")(attrition_dense_layer)
```

```python
# Create the model
model = Model(
    inputs=input_layer,
    outputs=[
        department_output_layer,
        attrition_output_layer
    ]
)

# Compile the model
model.compile(
    optimizer="adam",
    loss={
        "department_output": "categorical_crossentropy",
        "attrition_output": "categorical_crossentropy"
    },
    metrics={
        "department_output": "accuracy",
        "attrition_output": "accuracy"
    }
)

# Summarize the model
model.summary()
```

```python
# Train the model
_ = model.fit(
    X_train_scaled,
    {
        "department_output": y_train_department,
        "attrition_output": y_train_attrition
    },
    epochs=10,
    validation_data=(
        X_test_scaled,
        {
            "department_output": y_test_department,
            "attrition_output": y_test_attrition
        }
    )
)
```

```python
# Evaluate the model with the testing data
results = model.evaluate(
    X_test_scaled,
    {
        "department_output": y_test_department,
        "attrition_output": y_test_attrition
    }
)
results
```

```python
# Print the accuracy for both department and attrition
print("Department predictions accuracy:", results[3])
print("Attrition predictions accuracy:", results[4])
```
